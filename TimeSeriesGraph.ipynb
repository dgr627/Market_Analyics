```python


from xbbg import blp
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

vector = [".PBC G Index"]


data = blp.bdh(vector, 'LAST_PRICE', '20020101', datetime.date.today())
data["date"]=data.index
plt.figure(figsize=(10,6))
for col in vector:
    plt.plot(data.index, data[col], label=col[0:6])
    plt.xticks(rotation=45)
    plt.title("3m 10del Flies")
    plt.legend()
plt.show()
```

```python

vector = ["USDCNHV1M Curncy", "USDCNH25R1M Curncy", "USDCNH10B1M Curncy"]

data = blp.bdh(vector, 'LAST_PRICE', '20220501', datetime.date.today())
data["date"]=data.index
print(data)
plt.figure(figsize=(8,6))
plt.plot(data.index, data.iloc[:,0], label=vector[0])
plt.legend(loc='best')
plt.ylabel("Vol", labelpad=15)
plt.xticks(rotation=45)
plt.twinx()
plt.plot(data.index, data.iloc[:,1], label=vector[1], color='r')
plt.legend(loc='best')
plt.plot(data.index, data.iloc[:,2], label=vector[2], color='g')
plt.ylabel("Fly and RR", rotation=270, labelpad=15)
plt.title("USDCNH Vol vs Surface")
plt.legend(loc='lower right')
plt.show()
```
